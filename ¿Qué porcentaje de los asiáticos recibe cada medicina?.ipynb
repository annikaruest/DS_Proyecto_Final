# ¿Qué porcentaje de los asiáticos recibe cada medicina?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("diabetic_data.csv")
data.head()
data.info()

chinos = data[(data.race == "Asian")]

data.count()

a = chinos.count()
b =chinos[(chinos.insulin == "No")].count()
print ((a-b)/a*100) #48.05% recibieron insulina


b1 =chinos[(chinos.metformin == "No")].count()
print ((a-b1)/a*100) #17.94% recibieron metformin

b2 =chinos[(chinos.repaglinide == "No")].count()
print ((a-b2)/a*100) #2.65% recibieron repaglinide   

b3 =chinos[(chinos.nateglinide == "No")].count()
print ((a-b3)/a*100) #1.56% recibieron nateglinide

b4 =chinos[(chinos.chlorpropamide == "No")].count()
print ((a-b4)/a*100) #Nadie recibió chlorpropamide

b5 =chinos[(chinos.glimepiride == "No")].count()
print ((a-b5)/a*100) #3.90% recibieron glimepiride

b6 =chinos[(chinos.acetohexamide == "No")].count()
print ((a-b6)/a*100) #Nadie recibió acetohexamide

b7 =chinos[(chinos.glipizide == "No")].count()
print ((a-b7)/a*100) #13.73 recibieron glipizide

b8 =chinos[(chinos.glyburide == "No")].count()
print ((a-b8)/a*100) #10.76% recibieron glyburide

b9 =chinos[(chinos.tolbutamide == "No")].count()
print ((a-b9)/a*100) #Nadie recibio tolbutamide

b10 =chinos[(chinos.pioglitazone == "No")].count()
print ((a-b10)/a*100) #7.18% recibió pioglitazone

b11 =chinos[(chinos.rosiglitazone == "No")].count()
print ((a-b11)/a*100) #4.84% recibió rosiglitazone

b12 =chinos[(chinos.acarbose == "No")].count()
print ((a-b12)/a*100) #0.31% recibieron acarbose

b13 =chinos[(chinos.miglitol == "No")].count()
print ((a-b13)/a*100) #Nadie recibió miglitol

b14 =chinos[(chinos.troglitazone == "No")].count()
print ((a-b14)/a*100) #Nadie recibió troglitazone

b15 =chinos[(chinos.tolazamide == "No")].count()
print ((a-b15)/a*100) #NAdie recibió tolazamide

b16 =chinos[(chinos.examide == "No")].count()
print ((a-b16)/a*100) #Nadie recibió examide

b17 =chinos[(chinos.citoglipton == "No")].count()
print ((a-b17)/a*100) #Nadie recibió citoglipton
```
